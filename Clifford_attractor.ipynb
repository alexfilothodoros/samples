```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define equation
def clifford_attractor(x, y, a=-1.4, b=1.6, c=1.0, d=0.7):
    x_n1 = np.sin(a*y) + c*np.cos(a*x)
    y_n1 = np.sin(b*x) + d*np.cos(b*y)
    return x_n1, y_n1
```

We need to solve the following system of equations:

$x_{n+1}=\sin{ay_{n}} + c \cos{ax_{n}}
\\ 
y_{n+1}=\sin{bx_{n}} + d \cos{by_{n}}$,

where a, b, c, d are variables that define each attractor.

I am using the values from this source:  https://fronkonstin.com/2017/11/07/drawing-10-million-points-with-ggplot-clifford-attractors/



```python
# basic parameters
steps = 1000000
a, b, c, d =-1.24458046630025,-1.25191834103316, -1.81590817030519, -1.90866735205054
# initialize solutions arrays (+1 for initial conditions)
xx = np.empty((steps + 1))
yy = np.empty((steps + 1))
```

```python
# solve equation system
for i in range(steps):
    # Calculate derivatives	
    x_dot, y_dot = clifford_attractor(xx[i], yy[i],a=a,b=b,c=c,d=d)
    xx[i + 1] = x_dot
    yy[i + 1] = y_dot
    
```

```python
fig = plt.figure(facecolor='white', edgecolor='black')

plt.plot(xx, yy,'.',c='k',ms=0.1)
```

```python
bg_color = 'black'
fg_color = 'white'

fig = plt.figure(facecolor=bg_color, edgecolor=fg_color)
axes = fig.add_subplot(111)
axes.patch.set_facecolor(bg_color)
axes.xaxis.set_tick_params(color=fg_color, labelcolor=fg_color)
axes.yaxis.set_tick_params(color=fg_color, labelcolor=fg_color)
for spine in axes.spines.values():
    spine.set_color(fg_color)
    
plt.plot(xx, yy,'.',c='white',ms=0.1)
```
