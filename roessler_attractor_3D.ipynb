```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

We need to solve the following system of equations:

\begin{matrix}
\dot{x} = -y -z\\ 
\dot{y} = x + ay\\ 
\dot{z} = b + z(x-c)
\end{matrix}

,where a, b, c, d are the variables we can experiment with.

```python
# define equation
def rossler_attractor(x, y, z, a=0.1, b=0.1, c=14):
    x_dot = - y - z
    y_dot = x + a*y
    z_dot = b + z*(x - c)
    return x_dot, y_dot, z_dot

step_size = 0.01
steps = 100000
```

```python
xx = np.empty((steps + 1))
yy = np.empty((steps + 1))
zz = np.empty((steps + 1))

# initial conditions
xx[0], yy[0], zz[0] = (0.1, 0., 0.1)


# solve equation system
for i in range(steps):
    # Calculate derivatives
    x_dot, y_dot, z_dot = rossler_attractor(xx[i], yy[i], zz[i])
    xx[i + 1] = xx[i] + (x_dot * step_size)
    yy[i + 1] = yy[i] + (y_dot * step_size)
    zz[i + 1] = zz[i] + (z_dot * step_size)

```

```python

fig = plt.figure()
ax = fig.gca(projection='3d')
plt.gca().patch.set_facecolor('black')

plt.plot(xx, yy, zz,'.',color='white',ms=0.1)

# set limits
ax.set_xlim(-20,25)
ax.set_ylim(-25,20)
ax.set_zlim(0,50)

# remove ticks
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

ax.w_xaxis.set_pane_color((0.0, 0.0, 0.0, 1.0))
ax.w_yaxis.set_pane_color((0.0, 0.0, 0.0, 1.0))
ax.w_zaxis.set_pane_color((0.0, 0.0, 0.0, 1.0))

ax.view_init(30, -130)

plt.show()
```
